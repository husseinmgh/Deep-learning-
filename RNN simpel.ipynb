```python
from matplotlib import colors
import pandas as pd
import numpy as np
import keras
import keras.models
from datetime import datetime,date
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from keras.models import Sequential,load_model
from keras.layers import LSTM,Dropout,Dense,SimpleRNN
import tensorflow as tf

def dataparse(times):
    for i ,t in enumerate(times):
        print (i)
        times[i]=date(int(t[:4]),int(t[4:6]),int(t[6:8]))
    return times

data_file="ADA-USD"
df=pd.read_csv(data_file+".csv")
df["date"]=pd.to_datetime(df.Date,format="%Y-%m-%d")
df.index=df["date"]
print(df.head())
############################################
data=df.sort_index(ascending=True,axis=0)
dataset=data.values
time_step=60
rnn_units=50
lr=0.003
act_func="tanh" #sigmoid relu linear tanh
loss_fcn=keras.losses.MeanSquaredError()
ntrain=len(data)-300
feature_cols=[2,3]
target_cols=[2,3]
nfeature=len(feature_cols)

train_data=dataset[0:ntrain,feature_cols]
train_target=dataset[0:ntrain,target_cols]

test_data=dataset[ntrain-time_step:,feature_cols]
test_target=dataset[ntrain-time_step:,target_cols]

if(train_data.ndim<2):
    train_data=train_data.reshape(0,1)
    test_data=test_data.reshape(0,1)

if(True):
    x_scaler=MinMaxScaler(feature_range=(0,1))
    y_scaler=MinMaxScaler(feature_range=(0,1))
else:
    x_scaler=StandardScaler()
    y_scaler=StandardScaler()

train_data_scaled=x_scaler.fit_transform(train_data)
train_target_scaled=y_scaler.fit_transform(train_target)

x_train,y_train,x_test,y_test=[],[],[],[]

for i in range(time_step,len(train_data)):
    x_train.append(train_data_scaled[i-time_step:i,:])
    y_train.append(train_target_scaled[i,:])

x_train,y_train=np.array(x_train),np.array(y_train)

if(x_train.ndim<3):
    x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))


rnn_model=Sequential()

rnn_model.add(SimpleRNN(rnn_units,return_sequences=True,input_shape=(x_train.shape[1],x_train.shape[2]),activation=act_func))
rnn_model.add(SimpleRNN(rnn_units ,activation=act_func))
#rnn_model.add(keras.layers.BatchNormalization())
rnn_model.add(Dense(len(target_cols)))
rnn_model.summary()

opt=tf.keras.optimizers.Adam(learning_rate=lr)
rnn_model.compile(loss=loss_fcn,optimizer=opt)
#dense_model=load_model("dense_modelhw7.h5")
rnn_model.fit(x_train,y_train,epochs=1,batch_size=1,verbose=1)
rnn_model.save("Rnn_model.h5")

test_data_scaled=x_scaler.transform(test_data)
test_target_scaled=y_scaler.transform(test_target)

for i in range(time_step,len(test_data)):
    x_test.append(test_data_scaled[i-time_step:i,:])
    y_test.append(test_target_scaled[i,:])

x_test,y_test=np.array(x_test),np.array(y_test)

if(x_test.ndim<3):
    x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))


predicted_price=rnn_model.predict(x_test)
predicted_price=y_scaler.inverse_transform(predicted_price)

trn=data[0:ntrain]
tst=data[ntrain:]


plt.figure(figsize=(16,8))
tst['High-Pred']=predicted_price[:,0]
plt.plot(trn['High'])
plt.plot(tst['High'],'green',tst['High-Pred'],'red')
plt.title('High price history')

if (predicted_price.shape[1]>1):
    plt.figure(figsize=(16,8))
    tst['Low-Pred']=predicted_price[:,1]
    plt.plot(trn['Low'])
    plt.plot(tst['Low'],'blue',tst['Low-Pred'],'Orange')
    plt.title('Low price history')


plt.show()


x_test=x_test[0,:]
x_test=x_test.reshape(1,x_test.shape[0],x_test.shape[1])
for i in range(0,300):
    output=rnn_model.predict(x_test)
    predicted_price[i,:]=(y_scaler.inverse_transform(output))
    x_test[0,:-nfeature]=x_test[0,nfeature:]
    x_test[0,-nfeature:]=output

trn=data[0:ntrain]
tst=data[ntrain:]

tst['High-Pred']=predicted_price[:,0]
tst['Low-Pred']=predicted_price[:,1]

plt.figure(figsize=(16,8))
plt.plot(trn['High'])
plt.plot(tst['High'],'green',tst['High-Pred'],'red')
plt.title('High price history')

plt.figure(figsize=(16,8))
plt.plot(trn['Low'])
plt.plot(tst['Low'],'blue',tst['Low-Pred'],'Orange')
plt.title('Low price history')

plt.show()
# test_loss = rnn_model.evaluate(test_data_scaled, test_target_scaled)
```
